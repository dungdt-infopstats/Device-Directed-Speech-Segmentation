```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("tridungdo/ddsd-aug-ratio")

print("Path to dataset files:", path)
```

```python
list_file_id = list(df_test['file_id'].values)
```

```python
import os
import shutil
from tqdm import tqdm

root_path = os.path.join(path, 'concat_speech', 'augmented')
output_path = os.path.join('test')

for folder in tqdm(os.listdir(root_path), desc="Copying folders"):
    if folder in list_file_id:
        src = os.path.join(root_path, folder)
        dst = os.path.join(output_path, folder)
        if os.path.exists(src):
            shutil.copytree(src, dst, dirs_exist_ok=True)  # không báo lỗi nếu dst tồn tại
```
